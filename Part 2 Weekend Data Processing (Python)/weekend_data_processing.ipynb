## Importing Libraries

```python
import pandas as pd
```

## Reading the CSV Files

```python
saturday_data = pd.read_csv('data_2023-02-11.csv', delimiter = ';')
sunday_data = pd.read_csv('data_2023-02-12.csv', delimiter = ';')
```

```python
saturday_data
```

```python
sunday_data
```

## Combining Saturday and Sunday Data 

```python
weekend_data = pd.concat([saturday_data, sunday_data])
weekend_data.reset_index(drop = True, inplace = True)
```

```python
weekend_data["Generation Date"] = weekend_data.apply(lambda row: pd.Timestamp.now(), axis=1) 
```

```python
weekend_data
```

## Exporting the Dataframe back to CSV 

```python
weekend_data.to_csv("combined_weekend_data.csv", index=False)
```
